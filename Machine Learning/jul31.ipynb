```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# x = np.linspace(10,110,25).round()
# y = np.linspace(1000,10000,25).round()
x = np.random.randint(10,100,40)
x.sort()
y = np.random.randint(1000,10000,40)
y.sort()
data = {'A':x,'B':y}
```

```python
df = pd.DataFrame(data)
df
```

```python
plt.scatter(df['A'],df['B'])
```

```python
from sklearn.model_selection import train_test_split
a_train,a_test,b_train,b_test = train_test_split(df[['A']],df[['B']],test_size=0.3,random_state=101)
b_train.shape
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(a_train,b_train)
```

```python
prediction = reg.predict(a_test)
```

```python
plt.scatter(a_train,b_train)
```

```python
plt.scatter(a_test,b_test)
plt.plot(a_test,prediction)
```

```python
arr1 = np.array([1,2,3,4,5])
arr2 = np.array([5,7,9,11,15])
def gradient_descent(x,y):
    m_curr=0
    b_curr = 0
    iterations = 100
    learning_rate = 0.08
    n=len(x)
    plt.scatter(x,y)
    for i in range(iterations):
        y_pred = m_curr*x + b_curr
        plt.plot(x,y_pred,c='green')
        cost = (1/n)*sum([val**2 for val in y-y_pred])
        md = -(2/n)*sum(x*(y-y_pred))
        bd = -(2/n)*sum((y-y_pred))
        m_curr = m_curr - md*learning_rate
        b_curr = b_curr - bd*learning_rate

        print(f"m={m_curr}\t\tb={b_curr}\t\titeration={i}\t\tcost{cost}") 

gradient_descent(arr1,arr2)
```
