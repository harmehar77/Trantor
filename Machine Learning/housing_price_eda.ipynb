```python
data = df['Avg. Area Income']
plt.boxplot(data,vert=True,patch_artist=True)
mean_income = np.mean(data)
std_income = np.std(data)
threshold = 3
outliers = [x for x in data if np.abs(x - mean_income) > threshold * std_income]
print(len(outliers))
```

```python
X = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]
y = df[['Price']]
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
```

```python
scaled_data = scalar.fit_transform(X)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=1)
```

```python
scaled_data = pca.fit_transform(scaled_data)
```

```python
scaled_data
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(scaled_data,y,test_size=0.3,random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
predictions = reg.predict(x_test)
```

```python
plt.scatter(x_train,y_train)
```

```python
plt.scatter(predictions,y_test)
```

```python
#plt.scatter(x_test,y_test)
plt.scatter(y_test,predictions,lw=3,c='r')
```

```python
from sklearn import metrics
```

```python
income = df[['Avg. Area Income']]
```

```python
income_train,income_test = train_test_split(income,test_size=0.3,random_state=101)
```

```python
income_price = LinearRegression()
```

```python
income_price.fit(income_train,y_train)
```

```python
income_price.predict([[100000]])
```

```python
pred = income_price.predict(income_test)
```

```python
plt.scatter(income_train,y_train)
```

```python
#plt.scatter(income_test,y_test)
plt.scatter(y_test,pred,c='r')
```

```python
print(income_price.coef_)
print(income_price.intercept_)
```

```python
print(metrics.mean_absolute_error(income_test,pred))
```
