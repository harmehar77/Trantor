Correlation Methods()

1. Numpy: corrcoef(x,y)
corrcoef() returns the correlation matrix, which is a two-dimensional array with the correlation coefficients. the lower left and upper right values are equal and both represent the Pearson correlation coefficient for x and y. Eg:

x    1.00    0.76
y    0.76    1.00

2. Scipy: 
pearsonr(): r, p = scipy.stats.pearsonr(x, y)
spearmanr()
kendalltau()
these functions return objects that contain two values:
a. The correlation coefficient
b. The p-value

3. stats.correlation: returns person's correlation coefficient for two inputs

```python
print(cancer['DESCR'])
```

```python
df = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
df
```

```python
a = np.array(df['mean radius'])
a = (a - np.mean(a))/np.std(a)

b = np.array(df['mean texture'])
b = (b - np.mean(b))/np.std(b)

cov_mat = np.cov(a,b)
cov_mat
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
scalar.fit(df)
scaled_data = scalar.transform(df)
scaled_data
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(scaled_data)
reduced_data = pca.transform(scaled_data)
```
